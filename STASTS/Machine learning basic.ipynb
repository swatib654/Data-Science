```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Fix for full array printing (replace np.nan with sys.maxsize)
import sys
np.set_printoptions(threshold=sys.maxsize)

# Load Dataset
dataset = pd.read_csv(r"C:\Users\pc\Downloads\15th- SLR\House_data.csv")
space = dataset['sqft_living']
price = dataset['price']

x = np.array(space).reshape(-1, 1)
y = np.array(price)

# Corrected train_test_split import (from model_selection)
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=1/3, random_state=0)

# Fit Linear Regression Model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(xtrain, ytrain)

# Predict on Test Set
pred = regressor.predict(xtest)

# Visualize Training Set
plt.scatter(xtrain, ytrain, color='red')
plt.plot(xtrain, regressor.predict(xtrain), color='blue')
plt.title("Visuals for Training Dataset")
plt.xlabel("Space (sqft_living)")
plt.ylabel("Price")
plt.show()

# Visualize Test Set
plt.scatter(xtest, ytest, color='red')
plt.plot(xtrain, regressor.predict(xtrain), color='blue')  # Line from training set
plt.title("Visuals for Test Dataset")
plt.xlabel("Space (sqft_living)")
plt.ylabel("Price")
plt.show()
```
