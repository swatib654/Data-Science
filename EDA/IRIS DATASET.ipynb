```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you already have 'iris' DataFrame with 'Species' and 'PetalLengthCm'
# If not, uncomment and load manually:
# iris = pd.read_csv("your_iris_file.csv")

plt.figure(figsize=(8, 5))

# Plot boxplot and stripplot
ax = sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray", ax=ax)

# Colors for boxes (as many as categories in "Species")
colors = ['yellow', 'lightgreen', 'skyblue']

# Change the box colors using ax.patches
for i, patch in enumerate(ax.patches[:len(colors)]):  # Each box is a patch
    patch.set_facecolor(colors[i])
    patch.set_edgecolor('black')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Make sure only numeric columns are passed to .corr()
iris_numeric = iris.select_dtypes(include='number')

# Plot
fig = plt.gcf()
fig.set_size_inches(10, 7)

sns.heatmap(
    iris_numeric.corr(),  # only numeric data
    annot=True,
    cmap='cubehelix',
    linewidths=1,
    linecolor='k',
    square=True,
    vmin=-1,
    vmax=1,
    cbar_kws={"orientation": "vertical"},
    cbar=True
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
# Create a kde plot of sepal_length versus sepal width for setosa species of flower.
sub=iris[iris['Species']=='Iris-setosa']
sns.kdeplot(data=sub[['SepalLengthCm','SepalWidthCm']],cmap="plasma", shade=True, shade_lowest=False)
plt.title('Iris-setosa')
plt.xlabel('Sepal Length Cm')
plt.ylabel('Sepal Width Cm')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter only Iris-setosa
sub = iris[iris['Species'] == 'Iris-setosa']

# Create KDE plot for SepalLengthCm vs SepalWidthCm
sns.kdeplot(
    data=sub,
    x='SepalLengthCm',
    y='SepalWidthCm',
    fill=True,
    cmap='plasma',
    linewidths=1
)

plt.title('Iris-setosa KDE')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```
